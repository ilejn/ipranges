```python
import numpy as np
import pandas as pd
%pylab inline
```

```python
arr = np.loadtxt('stamps.dat')
```

```python
rarr = arr.reshape(int(len(arr)/5),5)
```

```python
frame = pd.DataFrame(rarr)
```

```python
frame.columns = ['ranges','STDMAP', 'BINSEARCH', 'UNORDEREDSET', 'SPARSEHASHSET']
frame.set_index('ranges', inplace=True)
frame
```

```python
figsize_=(10,10) 
style_=['x-', 'v-', 'D-', '+-']
frame.index.name = '# of ranges'
frame.columns.name = 'container'
```

```python
for lines in [3, 5, 6, 7]:  frame.head(lines).plot(figsize=figsize_, style=style_).set(ylabel='microseconds')
```
